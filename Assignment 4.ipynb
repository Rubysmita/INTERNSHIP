Q2.Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/. You need to find following details: A) Match title (I.e. 1st ODI) B) Series C) Place D) Date E) Time Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver=webdriver.Chrome(r"C:\chromedriver_win32 (1)\chromedriver.exe")
```

```python
driver.get(' https://www.bcci.tv/')
```

```python
seachinternational=driver.find_element_by_xpath("//li[@class='nav-item ml-0']/a")
driver.get(seachinternational.get_attribute('href'))
```

```python
try:
    format=driver.find_elements_by_xpath("//div[@class='cSBDisplay ng-binding']")
    format[1].click()
except:
    print("Error Occured")
```

```python
formatODI=driver.find_elements_by_xpath("//div[@class='cSBListItems ng-scope']")
formatODI[1].click()
```

```python
matchest=driver.find_elements_by_xpath("//span[@class='ng-binding']")
matches=[]
matches.append([i.text for i in matchest])
matches
```

```python
series=[]
series.extend([i.split(" ")[-1]for i in matches[0]])
series
```

```python
place=driver.find_elements_by_xpath("//div[@class='fix-place ng-binding ng-scope']")
venue=[]
venue.extend([i.text.split("-")[1] for i in place])
venue
```

```python
dateTag=driver.find_elements_by_xpath("//h5[@class='ng-binding']")
date=[]
date.extend([i.text for i in dateTag])
date
```

```python
timeTag=driver.find_elements_by_xpath("//h5[@class='text-right ng-binding']")
time=[]
time.extend([i.text for i in timeTag])
time
```

```python
df=pd.DataFrame()

df['Matches']=matches[0]

df['Series']=series

df['Place']=venue

df['date']=date

df['Time']=time
df
```

Q5.Scrape the details of trending repositories on Github.com. Url = https://github.com/ You have to find the following details: A) Repository title B) Repository description C) Contributors count D) Language used Note: - From the home page you have to click on the trending option from Explore menu through code

```python
driver=webdriver.Chrome(r"C:\chromedriver_win32 (1)\chromedriver.exe")
```

```python
driver.get('https://github.com/')
```

```python
try:
    driver.find_elements_by_xpath("//details[@class='HeaderMenu-details details-overlay details-reset width-full']")[1].click()
except:
    print("Error!!")
```

```python
try:
    driver.get(driver.find_elements_by_xpath("//a[@class='lh-condensed-ultra d-block no-underline position-relative Link--secondary py-2']")[10].get_attribute('href'))
except:
    print("Error!!")
```

```python
url=[]
try:
    urlTag=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
    url.extend([i.get_attribute('href') for i in urlTag])
except:
    print("Error!!")
```

```python
Repository_title=[]
Repository_description=[]
Contributors_count=[]
Language_used=[]
for i in url:
    driver.get(i)
    try:
        Repository_title.append(driver.find_element_by_xpath("//strong[@class='mr-2 flex-self-stretch']").text)
    except:
        Repository_title.append("-")
    try:
        Repository_description.append(driver.find_element_by_xpath("//div[@class='BorderGrid-cell']/p").text)
    except:
        Repository_description.append("-")
    try:
        Contributors_count.append(driver.find_elements_by_xpath("//h2[@class='h4 mb-3']/a/span")[-1].text)
    except:
        Contributors_count.append("-")
    try:
        Language_used.append(driver.find_element_by_xpath("//span[@class='color-fg-default text-bold mr-1']").text)
    except:
        Language_used.append("-")
```

```python
df=pd.DataFrame()
df['Repository_title']=Repository_title
df['Repository_description']=Repository_description
df['Contributors_count']=Contributors_count
df['Language_used']=Language_used
df
```

Q7.Scrape the details of Data science recruiters from naukri.com. Url = https://www.naukri.com/ You have to find the following details: A) Name B) Designation C) Company D) Skills they hire for E) Location Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
driver=webdriver.Chrome(r"C:\chromedriver_win32 (1)\chromedriver.exe")
```

```python
driver.get('https://www.naukri.com/data-science-jobs?src=gnbjobs_homepage_srch')
```

```python
driver.get(driver.find_element_by_xpath("//a[@title='Search Recruiters']").get_attribute("href"))
```

```python
driver.find_element_by_xpath("//input[@class='sugInp']").send_keys("Data Science")
```

```python
driver.find_element_by_xpath("//button[@class='fl qsbSrch blueBtn']").click()
```

```python
Name=[]
Designation=[]
Company=[]
Skills_hire=[]
Location=[]
try:
    nameTag=driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
    Name.extend([i.text for i in nameTag])
except:
    Name.append("-")
try:
    desTag=driver.find_elements_by_xpath("//span[@class='ellipsis clr']")
    Designation.extend([i.text for i in desTag])
except:
    Designation.append("-")

try:
    comTag=driver.find_elements_by_xpath("//a[@class='ellipsis']")[1::2]
    Company.extend([i.text for i in comTag])
except:
    Company.append("-")
    
try:
    skill=driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
    Skills_hire.extend([i.text for i in skill])
except:
    Skills_hire.append("-")

try:
    locTag=driver.find_elements_by_xpath("//small[@class='ellipsis']")
    try:
        Location.extend([i.text for i in locTag])
    except:
        Location.append("-")
except:
    Location.append("-")
```

```python
df=pd.DataFrame()
df["name"]=Name
df['Designation']=Designation
df['Company']=Company
df['Skills_hireFor']=Skills_hire
df
```
